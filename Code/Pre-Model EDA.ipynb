```python
# change the index and sort values to have individual lists of most used combinations

placement_1_list = placement_1_list.reset_index()
placement_1_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_1_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_2_list = placement_2_list.reset_index()
placement_2_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_2_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_3_list = placement_3_list.reset_index()
placement_3_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_3_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_4_list = placement_4_list.reset_index()
placement_4_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_4_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_5_list = placement_5_list.reset_index()
placement_5_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_5_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_6_list = placement_6_list.reset_index()
placement_6_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_6_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_7_list = placement_7_list.reset_index()
placement_7_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_7_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_8_list = placement_8_list.reset_index()
placement_8_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_8_list.sort_values(by = 'count', ascending=False, inplace = True)
```
